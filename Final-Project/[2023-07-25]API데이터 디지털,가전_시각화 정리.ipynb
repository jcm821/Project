# 데이터분석

* 클릭량과 매출액데이터를 비교했을때, 시즌성과 증가추세가 가장 뚜렸했던 디지털/가전과 출산/육아/반려용품 카테고리에 대해 세부분석을 진행함
* 특정 날짜 또는 기간에 클릭량이 증가하는 부분에 초점을 맞춰 분석 진행

# 1. 디지털/가전

* 특징이 유사한 항목끼리 결합하여 비교
1. **컴퓨터/주변기기**: PC, 노트북, PC부품, PC액세서리, 모니터, 모니터주변기기, 노트북액세서리, 소프트웨어, 주변기기, 멀티미디어장치, 저장장치
2. **가전/전자**: 계절가전, 생활가전, 영상가전, 멀티미디어, 이미용가전, 주방가전, 카메라/캠코더
3. **통신기기**: 음향가전, 태블릿PC, 태블릿액세서리
* 기타기기(자동차기기, 게임기/타이틀, 광학기기/용품, 학습기기)는 영향력이 미미하다 판단되어 제외

## 1. 분석 진행 과정
* 먼저 연령대별로 비교하고자 사용기기별, 성별에 따른 디지털/가전 카테고리 내 항목들이 시간에 따라 어떤 변화가 있었는지 시각화를 통해 파악 및 기초적인 인사이트 작성
* 이후 도출된 전체 시각화를 검토 후 연령대별로 특이점이 보이는 시각화 선정
* 20대의 컴퓨터 및 주변기기 클릭량의 경우가 가장 뚜렷한 변화를 보여 이 항목에 대해 포커스를 맞춰 세부적인 분석 및 인사이트 정리

### 시각화 결과 및 인사이트
- 20대에서 컴퓨터 및 주변기기 항목은 PC에서 검색량이 모바일의 2배정도로 매우 많음
- PC,모바일 공통으로 21년 8월, 22년 2월, 22년 8월에 검색량이 높은데, 매 학기가 시작되기 전에 필요한 제품들을 구비하는 것으로 추정
- 21년 11-12월, 22년 10월-11월에도 검색량이 많은데, 매년 11월마다 스마트스토어, 11번가, ssg, G마켓, 쿠팡등과 같은 e커머스 쇼핑몰을 포함하여 다양한 분야의 쇼핑몰에서 블랙프라이데이 할인행사를 진행함 → PC관련제품들도 할인하기 때문에 수요가 있다고 예상됨

### 20대 여성의 컴퓨터 및 주변기기 세부 분석

* 인사이트를 토대로 21년 11-12월, 특히 22년 10월-11월에 여성의 검색량이 많은것에 흥미를 가지게 되어 성비 데이터와 비교하여 인과관계 파악

* 예상했던 블랙프라이데이 이벤트와 연관이 있는지 22년 10월부터 12월까지 성별에 따른 컴퓨터 및 주변기기에 해당되는 품목들의 남여 검색성비를 비교했음
* **컴퓨터 및 주변기기에 해당하는 품목** : PC, PC부품, PC액세서리, 노트북, 멀티미디어장비, 모니터, 모니터주변기기, 소프트웨어, 저장장치, 주변기기
- 각 품목마다 22년10월1일부터 12월 31일 까지의 검색 성비를 보게 되면 컴퓨터/가전 카테고리 내에 어떤 품목이 여성의 클릭률이 급증했는지 알 수 있을것으로 예상
* 성별 그래프가 데칼코마니로 나오는 이유 → 전체가 100으로 남, 여가 양분되기 때문
- 성비그래프는 꺾은선 형태의 그래프로 도출했음(특정 날짜 사이를 비교하는 것이라 추세를 파악하기 용이할 것으로 생각해서)
- 왼쪽 눈금레이블은 여성기준, 오른쪽 눈금레이블은 남성기준이다.(태블로에서는 그래프에 커서를 대면 레이블 상관없이 비율이 나오나, 사진상에서는 혼동이 올 수 있으므로)

## 3. 결론

- 그래프에 날짜를 22년10월1일부터 12월 31일 까지로 필터링하여 도출한 결과임
- 20대 여성들이 디지털/가전 내의 카테고리 항목 전반적으로 11월과 12월에 클릭률이 남성에 비해 증가하는 구간이 자주 보임
- 특히 PC, PC부품, 모니터, 모니터주변기기, 저장장치 등의 항목은 이전에 여성들의 클릭률이 적었으나 11월,12월 사이에 상당히 높게 증가한 것으로 보아 블랙프라이데이와 연말(크리스마스)이벤트와 겹쳐 데스크탑(PC)에 대한 관심도가 증가한 것으로 보임
