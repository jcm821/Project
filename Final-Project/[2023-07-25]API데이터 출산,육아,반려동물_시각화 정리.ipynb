# 2. 출산/육아/반려용품

* API데이터의 경우, 출산/육아는 전체 항목, 반려동물에 대한 품목은 생활/가전에 존재
* 시간상 데이터를 합치진 못해서, 개별적인 시각화 자료로만 확인했음
* 디지철/가전과 같이 통합항목으로 본 것이 아니므로 그래프의 혼동을 방지하기 위해 PC+MO, 성별 전체 등의 항목 제외

### 출산/육아 중분류 시각화

* 기기별 출산/육아 카테고리 내 클릭량 시각화



* 성별에 따른 출산/육아 카테고리 내 클릭량 시각화



* 월별 전국 출생인구, 혼인인구 수



* 성별에 따른 출생/육아 카테고리 클릭량 시각화



### 시각화 결과

- 완구/매트, 유아용/잡화의 경우 12월에 검색량이 많음 → 크리스마스의 영향
- 성별 데이터와 비교했을때 남성은 분유, 완구/매트, 카시트 등의 항목에서 검색량이 많았고, 여성은 주로 수유용품, 외출용품, 이유식, 이유식용품 등에서 검색량이 많았다는 점에서 출산/육아를 준비하는 관점이 조금씩 다르다는 경향이 보임
- 다나와 기준으로 완구/매트안에 통상적으로 아는 장난감들/로봇/놀이매트 등이 포함되어 있으며, 유아잡화의 경우도 의류,신발 등이 포함된 항목임을 확인
- 기저귀, 수유용, 위생/건강용품, 이유식용품은 2~3월에 검색량이 많음 → 봄에 출산하는 사람들이 많음을 간접적으로 확인 가능 -> kosis 출산율 데이터를 통해 뒷받침

- 조출생률: 특정인구집단의 출산수준을 나타내는 기본적인 지표로서 1년간의 총 출생아수를 당해년도의 총인구로 나눈 수치를 1,000분비로 나타낸 것
- 전국 합계출산수만 추출해서 확인한 결과, 출산수가 꾸준히 감소하는게 보이지만, 매년 12월-1월에 출산 수가 급증하는 패턴이 보임 → 출산 이후 2-3월에 아기를 위한 용품(위에서 정의한 기저귀, 수유용, 위생/건강용품, 이유식용품등)의 검색량이 많아졌다는 것을 보조할 수 있음
- 매출액데이터와도 비교해보면 위의 출산데이터와 비슷한 패턴을 보임
- 대체적으로 모바일에서 검색량이 많았음, 앞선 항목들과 유사한 경향, 임산부의 움직임이 불편한 것을 생각하면 모바일에서 검색하고 구매하는 것이 비교적 편리할 것으로 예상


- 남성의 경우, 완구/매트, 유아동잡화, 카시트 순으로 검색량이 많았음
- 여성의 경우, 완구/매트, 유아동잡화, 이유식용품 순으로 검색량이 많았음
- 출산/육아를 준비하는 관점이 조금씩 다르다는 경향이 보임 → 육아를 준비하는데 있어 역할의 차이라고 생각됨

### 반려동물 용품에 대한 시각화

* 성별에 따른 생활/건강 카테고리 클릭량



### 시각화 결과
- 남성의 경우, 공구, 자동차용품, 주방용품 순으로 검색량이 많음
- 여성의 경우, 주방용품, 문구/사무용품, 반려동물 순으로 검색량이 많음
- 남성의 경우 자동차에 흥미를 가지며, 여성의경우 문구제품이나 반려동물에 흥미를 가진다는 점에서 남녀의 선호도가 다름을 알 수 있음
