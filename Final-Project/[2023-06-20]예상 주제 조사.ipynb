# 빅데이터를 이용한 철로사고방지 예측 서비스 

## 세부사항
철도 선로 상태 인식 데이터를 활용, 철로 선로 상태 데이터를 딥러닝에 학습시켜 철도 운영에서 발생할 수 있는 사고를 조기에 대응할 수 있도록 정보를 제공해주는 웹 서비스 구현

* 국내는 철도관제원이 CCTV모니터링하는 시스템
* 최초 주제를 구상했을 때 생각했던 부분은 딥러닝 구현을 통해 철도 운영에서 발생할 수 있는 사고예방에 필요한 정보를 웹 사이트에 구현
* 예를 들어, 레일의 어떤 부분이 정상 데이터와 비교했을 때 이상현상이라 판단되는 경우 사고발생 가능성이 높을 것으로 예상된다는 등의 정보를 사이트에 표시<br>
(나아가서 철로 곳곳 cctv를 설치하고 사이트와 연동하여 모니터링 시스템구축을 제안 해볼수도 있을 것 같음)
