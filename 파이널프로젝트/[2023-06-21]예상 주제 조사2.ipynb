# 추가 주제 선정

## 인삼 년근 및 등급 탐지 모델(Aihub)

### 세부사항
* 인삼을 년수에 따라 등급을 자동으로 식별하고 분류하여 실시간 인삼의 품질 서비스 제공 -> 웹사이트에 인삼의 품질정보 구현
* 기대효과로 출하가 집중되는 특정기간(6~8월)에 품질 검사 및 업무 효율성 증대시킬 수 있음
* 인삼부위에 대한 개념, 등급을 나누는 기준과 출하 프로세스 등에 대한 도메인 지식 학습 필요

### 데이터셋

* [인삼 년근 판독 및 등급 식별을 위한 인공지능 학습데이터] https://aihub.or.kr/aihubdata/data/view.do?currMenu=115&topMenu=100&aihubDataSe=realm&dataSetSn=71426

### 활용기술
* 데이터베이스 구축
* 데이터분석
* 딥러닝 -> Yolov5, EfficientNet B7를 이용하여 아래의 모델중에 하나 활용 가능(모델에 대한 지식 필요)
    - 인삼 부위 탐지 모델
    - 인삼 년근 및 등급 탐지 모델
    - 인삼 년근 및 등급 분류 모델
* 웹 사이트 구현

## 인공지능 기반 조식동물 및 해적생물 객체 탐지 모델(Aihub)

### 세부사항
* 데이터를 통해 딥러닝 학습 -> 생태환경에 피해를 유발하는 해양생물 7종을 분류하여 해당 피해생물의 정보를 웹 사이트에 구현 -> 어부들의 피해 감소 기대
* 피해유발 해양생물종에 대한 도메인지식 필요

### 데이터셋
* [연안어장 생태환경 피해 유발 해양생물 데이터셋] https://aihub.or.kr/aihubdata/data/view.do?currMenu=115&topMenu=100
* 추가적으로 필요한 부분은 공공데이터 CSV파일이나, open API활용

### 활용기술
* 데이터베이스 구축
* 데이터분석
* 딥러닝 -> Yolov7, Yolov7s, YoloR 등의 객체 탐지 모델 중에 하나를 활용하여 아래의 모델 활용
    - 인공지능 기반 조식동물 및 해적생물 객체 탐지 모델
* 웹 사이트 구현
